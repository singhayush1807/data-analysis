```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/t20.csv')

# Handle missing values (example)
df.fillna(method='ffill', inplace=True)


```

```python
df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting a bar chart for total goals scored per team
plt.figure(figsize=(10, 6))
sns.histplot(x='Inns', y='NO', data=df)
plt.title('Innings vs Not Out ')
plt.xticks(rotation=45)
plt.show()


```

```python
# Heatmap of correlations
p=pd.crosstab(df['4s'],df['6s'])
p
```

```python
sns.heatmap(p)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame with 'Mat', 'Runs', and 'category' columns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the line plot with better customization
sns.lineplot(x='Mat', y='Runs', data=df,  marker='o')

# Set the title and labels for axes
plt.title('Matches vs Runs', fontsize=16)
plt.xlabel('Matches', fontsize=14)
plt.ylabel('Runs Scored', fontsize=14)

# Improve legend positioning
plt.legend(title='Category', loc='upper left', fontsize=12)

# Display the plot
plt.show()

```

```python
df["Mat"].value_counts().plot(kind="pie", autopct="%.1f", figsize=(10, 10), fontsize=8)
```
